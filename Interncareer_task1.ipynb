```python
top_youtubers = df.sort_values(by='Comments', ascending=False)
top_youtubers.head(5)
```

```python
correlation_matrix = df[['Suscribers', 'Likes', 'Comments']].corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.show()
```

Here we can analyze the following things:

    1) as the subscribers increases the likes also increases but at a  slow rate
    2) as the subscribers increase the comments also increase but at very less rate
    3) the more number of likes increase with the number of subscribers.
    4) the most growth can be seen with increase in comments as the likes increases

```python
regional_categories = df.groupby(['Country', 'Categories']).size().unstack()
sns.heatmap(regional_categories, cmap='viridis', annot=True, fmt='d')
plt.show()
```

Hare we can see that the people in indonesia do not have a particular category of what they produce in their channels


```python
average_metrics = df[['Suscribers', 'Visits', 'Likes', 'Comments']].mean()
average_metrics.plot(kind='bar', figsize=(8, 6), colormap='viridis')
plt.ylabel('Mean')
plt.show()
```

```python
average_metrics = df[['Suscribers', 'Visits', 'Likes', 'Comments']].count()
average_metrics.plot(kind='bar', figsize=(8, 6), colormap='viridis')
plt.ylabel('Average Count')
plt.show()
```

```python
df['EngagementRate'] = (df['Likes'] + df['Comments']) / df['Visits']

# Plot engagement rate over time
plt.figure(figsize=(12, 6))
df['EngagementRate'].plot(title='Engagement Rate Over Time', xlabel='Date', ylabel='Engagement Rate')
plt.show()
```

```python
category_distribution = df['Categories'].value_counts()

# Plot the distribution of video categories
plt.figure(figsize=(10, 6))
category_distribution.plot(kind='bar', color='orange', alpha=0.8)
plt.title('Video Category Distribution')
plt.xlabel('Category')
plt.ylabel('Count')
plt.show()
```

```python
top_performers = df[df['Suscribers'] > df['Suscribers'].mean()]
top_performers
```

Here we can see that a total of 303 entries with subscribers more than the mean subscribers

```python
df['TotalEngagement'] = df['Likes'] + df['Comments'] + df['Suscribers']
top_5_youtubers = df.sort_values(by='TotalEngagement', ascending=False).head(5)
print(top_5_youtubers)
```
